# Iris Dataset Statistical Analysis Using NumPy

```python
import numpy as np

# Load CSV (skip header row)
iris = np.genfromtxt(
    "iris.csv",
    delimiter=",",
    skip_header=1,
    dtype=str
)

print(iris)
```

## Dataset Overview

```python
print("Dimensions:",iris.ndim)
print("Shape:",iris.shape)
print("Size:",iris.size)
print("Datatype:",iris.dtype)
```

##  Separate Features and Target Variable

```python
# Features (Id ko skip karke)
X = iris[:, 1:5].astype(float)

# Species
y = iris[:, 5]
```

##  Feature Extraction

```python
print("Features Shape :", X.shape)
print("Species Shape :", y.shape)

print("\nFirst 5 Feature Rows")
print(X[:5])

print("\nFirst 5 Species")
print(y[:5])
```

```python
sepal_length = X[:, 0]
sepal_width = X[:, 1]
petal_length = X[:, 2]
petal_width = X[:, 3]
print("Sepal Length :", sepal_length[:5])
print("Sepal Width :", sepal_width[:5])
print("Petal Length :", petal_length[:5])
print("Petal Width :", petal_width[:5])
```

##  Species Identification

```python
print("Unique Species:")
print(np.unique(y))
```

##  Boolean Indexing

```python
setosa = X[y == "Iris-setosa"]
versicolor = X[y == "Iris-versicolor"]
virginica = X[y == "Iris-virginica"]
```

```python
print("Setosa Shape :", setosa.shape)
print("Versicolor Shape :", versicolor.shape)
print("Virginica Shape :", virginica.shape)
```

```python
print(setosa[:3])
print()

print(versicolor[:3])
print()

print(virginica[:3])
```

##  Statistical Analysis

##  Species-wise Analysis

```python
print("Setosa Mean")
print(np.mean(setosa, axis=0))

print()

print("Versicolor Mean")
print(np.mean(versicolor, axis=0))

print()

print("Virginica Mean")
print(np.mean(virginica, axis=0))
```

```python
print(np.median(setosa, axis=0))

print(np.median(versicolor, axis=0))

print(np.median(virginica, axis=0))
```

```python
print(np.std(setosa, axis=0))

print(np.std(versicolor, axis=0))

print(np.std(virginica, axis=0))
```

```python
print(np.min(setosa, axis=0))

print(np.min(versicolor, axis=0))

print(np.min(virginica, axis=0))
```

```python
print(np.max(setosa, axis=0))

print(np.max(versicolor, axis=0))

print(np.max(virginica, axis=0))
```

```python
print("="*50)
print("OVERALL DATASET ANALYSIS")
print("="*50)

print("Mean :", np.mean(X, axis=0))
print("Median :", np.median(X, axis=0))
print("Standard Deviation :", np.std(X, axis=0))
print("Minimum :", np.min(X, axis=0))
print("Maximum :", np.max(X, axis=0))
```

##  Maximum and Minimum Analysis

```python
max_index = np.argmax(sepal_length)

print("Flower with Maximum Sepal Length")
print("Index :", max_index)
print("Species :", y[max_index])
print("Sepal Length :", sepal_length[max_index])
```

```python
min_index = np.argmin(petal_width)

print("Flower with Minimum Petal Width")
print("Index :", min_index)
print("Species :", y[min_index])
print("Petal Width :", petal_width[min_index])
```

```python
long_sepal = X[sepal_length > 7]

print("Flowers having Sepal Length > 7")
print(long_sepal)
print("Count :", len(long_sepal))
```

## Random Sampling

```python
np.random.seed(42)

sample = np.random.choice(len(X), 5, replace=False)

print("Random Flower Indexes")
print(sample)

print("\nRandom Flower Data")
print(X[sample])

print("\nSpecies")
print(y[sample])
```

##  Sorting Feature Values

```python
sorted_sepal = np.sort(sepal_length)

print("Smallest 10 Sepal Lengths")
print(sorted_sepal[:10])

print()

print("Largest 10 Sepal Lengths")
print(sorted_sepal[-10:])
```

##  Key Insights

```python
print("="*60)
print("IRIS DATASET INSIGHTS")
print("="*60)

print("1. Total Flowers :", len(X))
print("2. Total Features :", X.shape[1])
print("3. Species :", np.unique(y))

print("4. Average Sepal Length :", np.mean(sepal_length))
print("5. Average Petal Length :", np.mean(petal_length))

print("6. Maximum Sepal Length :", np.max(sepal_length))
print("7. Minimum Sepal Length :", np.min(sepal_length))

print("8. Maximum Petal Width :", np.max(petal_width))
print("9. Minimum Petal Width :", np.min(petal_width))
```
